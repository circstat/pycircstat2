```python
import numpy as np

from pycircstat2 import Circular, data
```

## Circular Data Examples

### Ungrouped Data / Scatter Diagram

```python
data_zar_ex2_ch26 = data('zar_ex2_ch26')
data_zar_ex2_ch26
```

```python
circ_zar_ex2_ch26 = Circular(data=data_zar_ex2_ch26['data'], 
                             unit=data_zar_ex2_ch26['unit'],
                                k=data_zar_ex2_ch26['k'])
circ_zar_ex2_ch26
```

```python
circ_zar_ex2_ch26.plot()
```

### Grouped Data / Rose Diagram

```python
data_zar_ex3_ch26 = data('zar_ex3_ch26')
data_zar_ex3_ch26
```

```python
circ_zar_ex3_ch26 = Circular(data=data_zar_ex3_ch26['data'], 
                                w=data_zar_ex3_ch26['w'],
                             unit=data_zar_ex3_ch26['unit'],
                                k=data_zar_ex3_ch26['k'])
circ_zar_ex3_ch26
```

```python
circ_zar_ex3_ch26.plot()
```
