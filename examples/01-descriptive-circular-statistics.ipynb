```python
import numpy as np
import matplotlib.pyplot as plt

from pycircstat2 import Circular, load_data
```

## Circular Data Examples

### Ungrouped Data / Scatter Diagram

```python
data_zar_ex2_ch26 = load_data('D1', source='zar_2010', print_meta=True)
```

```python
data_zar_ex2_ch26
```

```python
circ_zar_ex2_ch26 = Circular(data=data_zar_ex2_ch26['θ'].values, 
                             unit='degree',
                             n_intervals=360)
circ_zar_ex2_ch26
```

```python
fig = plt.figure(figsize=(24, 12))
ax = [fig.add_subplot(121 + i, projection='polar') for i in range(2)]
circ_zar_ex2_ch26.plot(density_kwargs={'method': 'nonparametric'}, ax=ax[0])
circ_zar_ex2_ch26.plot(density_kwargs={'method': 'MoVM'}, ax=ax[1])
```

### Grouped Data / Rose Diagram

```python
data_zar_ex3_ch26 = load_data('D2', source='zar_2010', print_meta=True)
```

```python
data_zar_ex3_ch26
```

```python
circ_zar_ex3_ch26 = Circular(data=data_zar_ex3_ch26['θ'].values, 
                           w=data_zar_ex3_ch26['w'].values,
                           unit='degree',
                           n_intervals=360)
circ_zar_ex3_ch26
```

```python
circ_zar_ex3_ch26.plot(density_kwargs={'method': 'MoVM'})
```
