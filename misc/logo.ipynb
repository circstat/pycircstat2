```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import ticker
from pycircstat2.distribution import jonespewsey_asymext
```

```python
n = 200
x = np.linspace(0, 2 * np.pi, n)
mu = np.pi / 2

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(111, polar=True)

for nu in np.linspace(0, 0.9999, 12):
    ax1.plot(x, jonespewsey_asymext.pdf(x, kappa=5, psi=.5, nu=nu, xi=mu) + 1, linestyle='-', color='black', alpha=0.1)

psis = [0]
for j, ax in enumerate([ax1]):
    rtick = [0, 1]
    ax.spines["polar"].set_visible(False)
    ax.set_rgrids(rtick, ["" for _ in range(len(rtick))], fontsize=16)
    gridlines = ax.yaxis.get_gridlines()
    gridlines[-1].set_color("k")
    gridlines[-1].set_linewidth(1)

    position_major = []
    ax.xaxis.set_major_locator(ticker.FixedLocator(position_major))

    ax.text(x=0, y=0, s='PyCircStat', ha='center', va="center", color="C0", fontsize=50, fontweight='bold', zorder=1)
    ax.text(x=0, y=0, s='2', ha='center', va="center", color="C0", fontsize=75, fontweight='bold', zorder=0, alpha=.5)

    ax.set_ylim(0, 2)

    if j == 2:
        ax.legend(frameon=False)

fig.subplots_adjust(wspace=0., hspace=0, left=-1.25, top=0.15,right=0.,bottom=0.)
fig.tight_layout(pad=0)

fig.savefig('./logo.png', transparent=True)

# fig.show()
```
